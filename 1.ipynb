```python
# prediction house prices in Boston using the boston dataset in sklearn
from sklearn import datasets
boston = datasets.load_boston()
boston # given in dictionary form
```

```python
type(boston) # some sort of object in sklearn
```

```python
X = boston.data
Y = boston.target
print(X.shape)
print(Y.shape)
print(type(X))
print(type(Y))
```

X : 506 data points (rows) and 13 features
Y : target column of 506 rows

```python
import pandas as pd
df = pd.DataFrame(X)
df.describe()
```

```python
# Now let's train the algorithm ... here we are gonna use linear regression
# split the dataset into two parts : random split
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
print(X_train.shape)
print(X_test.shape) # 1 : 3 split by default
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression # the LinearRegression class is imported
alg1 = LinearRegression() # this gives us the algorithm object

#hey, algorithm, learn from this dataset
alg1.fit(X_train, Y_train)
```

```python
# lets test
Y_pred = alg1.predict(X_test)
Y_pred
```

```python
#compare Y_pred and Y_test
#lets plot
import matplotlib.pyplot as plt
plt.scatter(Y_pred,Y_test)
plt.axis([0, 40, 0, 40])
# ideally the points should be on the y = x line
plt.show()
```
