Create two arrays, x and y.

x should hold first 20 even points starting from 0 and y should be an exponentiation array where y[i] = 2^x[i].
Plot a line graph, line should be of color blue and it should be a dashed line (like -----).
Find the value of x from where there is slight increase in value of y as shown in your plot.
Plot and print the value of x.

Note: Here x must be multiple of 5.

Output Format

x



```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 40, 2, dtype = 'int64') # int32 isn't appropriate here coz max value of int32 is (2^31) - 1 [signed integer]
y = 2 ** x
plt.plot(x, y, 'b--')
plt.grid()
plt.show()
y
```

### Customizing the graph

```python
x = np.arange(0, 5, 0.1)
y = x ** 2
y2 = x ** 3
plt.plot(x, y, color = 'Blue', linewidth = 5, label = 'x^2')
plt.plot(x, y2, color = 'Green', marker = 'o', label = 'x^3')
plt.ylabel('y')
plt.xlabel('x')
plt.axis([0, 10, 0, 150])
plt.grid()
plt.legend()
plt.title('matplotlib demo')
plt.text(1.5, 60, 'Hello Babes!', fontsize = 18)
plt.show()
```

### Bubble Charts


```python
import matplotlib.pyplot as plt
import numpy as np
year = [2014, 2015, 2016, 2017, 2018]
salary = [20, 25, 28, 30, 31]
population = [200, 230, 250, 300, 370]

c = np.arange(0, len(year))
plt.title('Bubble Chart')
plt.scatter(year, salary, s = population, c = c, alpha = 0.5, marker = '^', edgecolors = 'black')

for i in range(len(year)):
    plt.text(year[i] + 0.1 , salary[i] + 0.1, population[i])
    
plt.show()
```

We are given data of Microsoft Corporations, it contains the gross annual revenue in billion U.S dollar and the number of employees in thousands and year. Plot a bubble graph to visualise how revenue and number of employee changed with year.

Find years where there is a drastic increase in gross revenue of Microsoft Corporations (from previous and next year both). Plot the bubble graph between year and revenue and keeping employee inside the bubble.
Print the year, revenue and number of the employee where there is a drastic increase in revenue ( Top 2 ).

@Note:For finding the year where there is a drastic increase in gross revenue check whether the revenue of that year is greater than in previous years and greater than next year. Here years should be printed in ascending order.

Output Format:


year1 revenue1 employee1

year2 revenue2 employee2

. . . 

. . .

. . .



```python
import numpy as np
import matplotlib.pyplot as plt
# given data
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
employees = np.array(employees)
year = np.array(year)
revenue = np.array(revenue)
# revenue and no. of employees vs year
color = np.arange(0, len(year))
plt.scatter(year, revenue, marker = '^', s = employees, c = color, edgecolors = 'black')
for i in range(len(year)):
    plt.text(year[i] + 0.1, revenue[i] + 0.1, employees[i] )
plt.axis([2004, 2019, 0, 120])
plt.show()

# first let's get the index of the spikes (as demanded in the problem statement)


```
